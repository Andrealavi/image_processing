```python
import numpy as np
import numpy.typing as npt
from skimage import io
import matplotlib.pyplot as plt
```

```python
img = io.imread("./img/original_lena.gif")
img = np.reshape(img, shape=(512, 512))

plt.imshow(img, cmap="gray")
plt.show()
```

```python
x, y = np.meshgrid(np.linspace(-1,1, 512), np.linspace(-1,1, 512))

dst = np.sqrt(x**2 + y**2)
sigma = 0.02
muu = 0

normal = 1/(2.0 * np.pi * sigma**2)

gauss = np.exp(-((dst-muu)**2 / (2.0 * sigma**2))) * normal

plt.imshow(gauss, cmap="binary")
plt.show()
```

```python
x, y = np.mgrid[0:512, 0:512]
centered_img = img * (-1)**(x+y)

img_dft = np.fft.fft2(centered_img)

filtered_img = np.fft.ifft2(img_dft*gauss)

filtered_img = filtered_img * (-1)**(x+y) 

plt.imshow(filtered_img.real, cmap="gray")
plt.show()
```
